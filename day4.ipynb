```python
import cv2
import matplotlib.pyplot  as plt 
import numpy as np 
import requests
from PIL import Image

car_cascade = cv2.CascadeClassifier("cars.xml")

img = Image.open(requests.get('https://media.istockphoto.com/id/155287967/photo/cars-in-rush-hour-with-traffic-at-dawn.jpg?s=612x612&w=0&k=20&c=tDAaJQMHIh6nFDr0rJlD44tEtmj2srdpoSTKL7C5SGE=', stream=True).raw)

img = img.resize((450,250))

img_arr = np.array(img)

gray = cv2.cvtColor(img_arr, cv2.COLOR_BGR2GRAY)

cars = car_cascade.detectMultiScale(gray, 1.1,1)

car_count = 0
for (x,y,w,h) in cars:
    cv2.rectangle(img_arr, (x,y), (x+w, y+h), (0,0,255),2)
    car_count += 1

plt.imshow(cv2.cvtColor(img_arr, cv2.COLOR_BGR2RGB))
plt.title("car detection")
plt.show()

print("عدد السيارات المكتشفة", car_count)


```

```python

import cv2

# تعليقات لشرح المراحل المختلفة في الكود

# تحميل ملف تعريف الوجه
face_detector = cv2.CascadeClassifier('haarcascade_frontalcatface.xml')

# فتح ملف الفيديو
cape = cv2.VideoCapture('vid.mp4')

# التحقق من نجاح فتح ملف الفيديو
if not cape.isOpened():
    print("تعذر فتح ملف الفيديو")
    exit()

# إعداد متغيرات للون وحجم المربع المستخدم في تحديد الوجوه
color = (255, 0, 0)  # اللون الأزرق
thickness = 5  # سمك الخط

# حلقة الفيديو
while True:
    # قراءة الإطار التالي من الفيديو
    ret, img = cape.read()

    # التحقق من نجاح قراءة الإطار
    if not ret:
        print("تعذر قراءة الإطار")
        break

    # تحويل الإطار إلى grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # اكتشاف الوجوه في الإطار
    faces = face_detector.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=4)

    # رسم مربع حول كل وجه مكتشف
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), color, thickness)

    # عرض الإطار الحالي
    cv2.imshow('Video',img)

    # إضافة تأخير طفيف (ضبط حسب الحاجة)
    cv2.waitKey(30)

    # التحقق من الضغط على 'ESC' للخروج
    key = cv2.waitKey(1) & 0xFF
    if key == 27:  # رمز ASCII لـ 'ESC'
        break

# إغلاق ملف التقاط الفيديو وإغلاق نافذة العرض
cape.release()
cv2.destroyAllWindows()
```

```python
face_detector = cv2.CascadeClassifier('haarcascade_frontalcatface.xml')

cape = cv2.VideoCapture('vid.mp4')

if not cape.isOpened():
    print("تعذر فتح الفيديو")
    exit()

color = (255,0,0)
thickness = 5

while True:
    ret, img = cape.read()

    if not ret:
        print("تعذر قراءة الاطار")
        break

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    faces = face_detector.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=4)

    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w, y+h), color, thickness)

    cv2.imshow('Video', img)

    cv2.waitKey(30)    

    key = cv2.waitKey(1) & 0xFF

    if key == 27:
        break

cape.release()
cv2.destroyAllWindows()
```
